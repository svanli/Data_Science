### Two Interfaces for the Price of One

#### MATLAB-style Interface

```python
plt.figure()   # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

#### Object-oriented interface

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

### Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color = 'blue')          # specify color by name
plt.plot(x, np.sin(x - 1), color = 'g')             # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.75')          # grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color = '#FFDD44')       # hex code (RRGGBB, 00 to FF)
plt.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color = 'chartreuse');   # HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle = 'solid')
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle = '-')  # solid
plt.plot(x, x + 5, linestyle = '--') # dashed
plt.plot(x, x + 6, linestyle = '-.') # dashdot
plt.plot(x, x + 7, linestyle = ':'); # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashed black
plt.plot(x, x + 3, ':r'); # dotted red
```

### Continuous Errors

```python
from sklearn.gaussian_process import GaussianProcessRegressor

# Define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, dyfit = gp.predict(xfit[:, np.newaxis], return_std = True)

# Visualize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color = 'gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                 color = 'gray', alpha = 0.2)
plt.xlim(0, 10);
```

### Density and Contour Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### Visualizing a Three-Dimensional Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors = 'black');
```

```python
plt.contour(X, Y, Z, 20, cmap = 'RdGy'); # RdGy means Red-Grey colormap
```

```python
plt.contourf(X, Y, Z, 20, cmap = 'RdGy')
plt.colorbar();
```

```python
plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'lower', cmap = 'RdGy',
           interpolation = 'gaussian', aspect = 'equal')
plt.colorbar();
```

```python
contours = plt.contour(X, Y, Z, 3, colors = 'black')
plt.clabel(contours, inline = True, fontsize = 8)

plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'lower',
           cmap = 'RdGy', alpha = 0.5)
plt.colorbar();
```

### Histograms, Binnings, and Density

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
rng = np.random.default_rng(1701)
data = rng.normal(size = 1000)

plt.hist(data);
```

```python
plt.hist(data, bins = 30, density = True, alpha = 0.5,
         histtype = 'stepfilled', color = 'steelblue',
         edgecolor = 'none');
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

kwargs = dict(histtype = 'stepfilled', alpha = 0.3, density = True, bins = 40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
counts, bin_edges = np.histogram(data, bins = 5)
print(counts)
```

### Two-Dimensional Histograms and Binnings

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10_000).T
```

#### <code>plt.hist2d</code>: Two-Dimensional Histogram

```python
plt.hist2d(x, y, bins = 30)
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins = 30)
print(counts.shape)
```

#### <code>plt.hexbin</code>: Hexagonal Binnings

```python
plt.hexbin(x, y, gridsize = 30)
cb = plt.colorbar(label = 'count in bin')
```

#### Kernel Density Estimation

```python
from scipy.stats import gaussian_kde

# Fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# Evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
           origin = 'lower', aspect = 'auto',
           extent = [-3.5, 3.5, -6, 6])
cb = plt.colorbar()
cb.set_label('density')
```
