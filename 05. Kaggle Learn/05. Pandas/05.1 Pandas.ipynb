```python
review_points_mean = reviews['points'].mean()
reviews['points'].map(lambda p: p - review_points_mean)
```

```python
help(map)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis = 'columns').head(3)
```
