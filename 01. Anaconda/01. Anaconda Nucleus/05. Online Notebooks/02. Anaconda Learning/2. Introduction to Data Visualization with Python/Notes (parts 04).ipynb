```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset("tips")

# Create a new Figure and Axes
fig, ax = plt.subplots()

# Use the axes-level function to create a plot
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)

# Customize the plot further using Matplotlib's Axes methods
ax.set_title("Tips vs Total Bill")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

# Display the plot
plt.show()
```

```python
fig, ax = plt.subplots()

#sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)

tips[tips.sex=='Female'].plot(x="total_bill", y="tip", kind='scatter', ax=ax, c='r', label='Female')
tips[tips.sex=='Male'].plot(x="total_bill", y="tip", kind='scatter', ax=ax, c='b', label='Male')

ax.set_title("Tips vs Total Bill")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

```python
fig, ax = plt.subplots()

#sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)

tips[tips.smoker=='No'].plot(x="total_bill", y="tip", kind='scatter', ax=ax, c='r', label='No smoker')
tips[tips.smoker=='Yes'].plot(x="total_bill", y="tip", kind='scatter', ax=ax, c='b', label='Smoker')

ax.set_title("Tips vs Total Bill")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

```python
fig, ax = plt.subplots()

#sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)

tips[tips.day=='Sat'].plot(x="total_bill", y="tip", kind='scatter', ax=ax, c='r', label='Sat')
tips[tips.day=='Sat'].plot(x="total_bill", y="tip", kind='scatter', c='b', label='Sun')
tips[tips.day=='Thur'].plot(x="total_bill", y="tip", kind='scatter', c='orange', label='Thur')
tips[tips.day=='Fri'].plot(x="total_bill", y="tip", kind='scatter', c='g', label='Fri')

ax.set_title("Tips vs Total Bill")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

Use Seaborn to visualize the distribution of body mass, the relationship between flipper length and body mass, and visualize the differences among species.

1. body_mass_g distribution

```python
sns.histplot(
    data=dfe, x='body_mass_g', kde=True, hue='species'
).set(
    title='Body mass distribution', xlabel='Body mass in g'
)

plt.show()
```

2. flipper_length_mm vs body_mass_g

```python
sns.scatterplot(
    data=dfe, x='body_mass_g', y='flipper_length_mm'
).set(
    title='Body mass vs flipper length', xlabel='Body mass in g', ylabel='Flipper length in mm')

plt.show()
```

3. flipper_length_mm vs body_mass_g vs species

```python
sns.scatterplot(
    data=dfe, x='body_mass_g', y='flipper_length_mm', hue='species'
).set(
    title='Body mass vs flipper length', xlabel='Body mass in g', ylabel='Flipper length in mm')

plt.show()
```
