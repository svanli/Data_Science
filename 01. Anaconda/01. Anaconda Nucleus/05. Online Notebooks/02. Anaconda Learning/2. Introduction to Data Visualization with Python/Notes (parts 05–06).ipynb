```python
dfe = pd.read_csv('penguins.csv')
```

```python
dfe.head()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(7, 4), constrained_layout=True)

ax4.remove()

# Matplotlib
ax1.hist(dfe['body_mass_g'])
ax1.set_xlabel('Body mass in g')
ax1.set_ylabel('Frequency')

# pandas
dfe['body_mass_g'].plot(kind='hist', ax=ax2)
ax2.set_xlabel('Body mass in g')

# Seaborn
sns.histplot(data=dfe, x='body_mass_g', kde=True, ax=ax3)
ax3.set_xlabel('Body mass in g')
ax3.set_ylabel('Frequency')

# Title
fig.suptitle('Distribution of body mass', fontsize=16)

plt.show()
```

```python
fig, axs = plt.subplot_mosaic("""
    ABB
    CBB
    """, figsize=(7, 4), layout='constrained')

# Matplotlib
axs['A'].hist(dfe['body_mass_g'])
axs['A'].set_xlabel('Body mass in g')
axs['A'].set_ylabel('Frequency')

# pandas
dfe['body_mass_g'].plot(kind='hist', ax=axs['B'])
#axs['B'].set_xlabel('Body mass in g')

# Seaborn
sns.histplot(data=dfe, x='body_mass_g', kde=True, ax=axs['C'])
axs['C'].set_xlabel('Body mass in g')
axs['C'].set_ylabel('Frequency')

# Title
fig.suptitle('Distribution of body mass', fontsize=16)

plt.show()
```

```python
dfe.groupby('species')['body_mass_g'].mean()
```

```python
fig, ax = plt.subplots()

sns.barplot(data=dfe, x='species', y='body_mass_g', ax=ax)

ax.set_title('Body mass per species with averages')

ax.annotate(text='average\n3 701', xy=(0, 4500),
            horizontalalignment='center', verticalalignment='bottom', #color='b',
            bbox=dict(boxstyle='round,pad=0.1', fc='w', lw=7))

ax.annotate(text='average\n3 733', xy=(1, 4500),
            horizontalalignment='center', verticalalignment='bottom',
            bbox=dict(boxstyle='round,pad=0.1', fc='w', lw=7))

ax.annotate('Maximum', xy=(2, 6500), xytext=(2, 8000),
            horizontalalignment='center', verticalalignment='bottom',
            bbox=dict(boxstyle='round,pad=0.1', fc='w', lw=7),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate(text='average\n5 076', xy=(2, 5500),
            horizontalalignment='center', verticalalignment='bottom',
            bbox=dict(boxstyle='round,pad=0.1', fc='w', lw=7))

ax.set_xlim(-1, 3)
ax.set_ylim(2000, 9000)

ax.set_xlabel(None)
ax.set_ylabel('Body mass in g')


plt.show()
```
