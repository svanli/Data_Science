# 2. How to read a plot

```python
import matplotlib.pyplot as plt
```

```python
figsize = (14, 5)
```

```python
print(figsize)
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
```

```python
plt.plot(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot([1, 2, 3, 4], [1, 4, 2, 3])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```
