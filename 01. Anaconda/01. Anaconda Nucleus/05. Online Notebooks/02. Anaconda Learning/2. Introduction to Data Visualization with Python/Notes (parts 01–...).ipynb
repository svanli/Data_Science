```python
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot([1, 2, 3, 4], [1, 4, 2, 3])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

plt.plot(
    [1, 2, 3, 4], [1, 4, 2, 3], 
    color = 'red', 
    marker = '.',
    linestyle = 'dashed',
    markersize = 5,
    linewidth = 1
)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot(
    [1, 2, 3, 4], [1, 4, 2, 3], 
    label = 'FY2223',
    color = 'blue', 
    marker = '.',
    markersize = 5,
    linewidth = 1
)

ax.plot(
    [1, 2, 3, 4], [1, 3, 3, 2], 
    label = 'FY2122',
    color = 'grey', 
    marker = '.',
    linestyle = 'dashed',
    markersize = 5,
    linewidth = 1
)

ax.bar(
    [1, 2, 3, 4], [0.3, 0.5, 0.1, 0.4],
    label = 'ABO sales'
)

ax.set_xlabel('Quarter')
ax.set_ylabel('mln â‚¬')
ax.set_title('Sales in FY2223 and FY2122')
ax.legend()

plt.show()
```
