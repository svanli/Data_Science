# 2. How to read a plot

```python
import matplotlib.pyplot as plt
```

```python
figsize = (14, 5)
```

```python
print(figsize)
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
```

```python
plt.plot(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot([1, 2, 3, 4], [1, 4, 2, 3])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

plt.scatter([1, 2, 3, 4], [1, 4, 2, 3])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

plt.bar([1, 2, 3, 4], [1, 4, 2, 3])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

plt.plot(
    [1, 2, 3, 4], [1, 4, 2, 3], 
    color = 'red', 
    marker = '.',
    linestyle = 'dashed',
    markersize = 5,
    linewidth = 1
)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('My Plot')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

plt.plot(
    [1, 2, 3, 4], [1, 4, 2, 3], 
    color = 'blue', 
    marker = '.',
    markersize = 5,
    linewidth = 1
)

plt.plot(
    [1, 2, 3, 4], [1, 3, 3, 2], 
    color = 'grey', 
    marker = '.',
    linestyle = 'dashed',
    markersize = 5,
    linewidth = 1
)

ax.set_xlabel('Quarter')
ax.set_ylabel('mln â‚¬')
ax.set_title('Sales in FY2223 and FY2122')

plt.show()
```
