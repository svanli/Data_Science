```python
import pandas as pd
import numpy as np

# Sample data
data = {
    "Student": ["Alice", "Bob", "Alice", "Bob", "Alice", "Bob"],
    "Subject": ["Math", "Math", "Science", "Science", "Math", "Science"],
    "Score": [90, 85, 78, 92, 88, 80]
}

df = pd.DataFrame(data)

# Group by 'Subject' and calculate average score normalized by standard deviation
grouped = df.groupby("Subject").aggregate(
    {
        "Score": lambda scores: np.mean(scores) / np.std(scores)
    }
)

print(grouped)
```

```python
import pandas as pd
import numpy as np

# Sample data
data = {
    "Student": ["Alice", "Bob", "Alice", "Bob", "Alice", "Bob"],
    "Subject": ["Math", "Math", "Science", "Science", "Math", "Science"],
    "Score": [90, 85, 78, 92, 88, 80]
}

df = pd.DataFrame(data)

# Group by 'Subject' and calculate normalized average score by standard deviation
grouped = df.groupby("Subject").aggregate(
    {
        "Score": lambda x: (np.mean(x) / np.std(x))
    }
)

print(grouped)
```
