## Derivatives

```python
y2 = 4.41
y1 = 4.0
x2 = 2.1
x1 = 2.0

m = (y2 - y1) / (x2 - x1)

print(round(m, 1))
```

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2

slope_at_2  = derivative_x(my_function, 2, .00001)

print(slope_at_2)
```

## The Chain Rule

```python
from sympy import *
```

```python
x = symbols('x')

z = (x**2 + 1)**3 - 2

dz_dx = diff(z, x)

print(dz_dx)
```

```python
x, y = symbols('x y')

_y = x**2 + 1
dy_dx = diff(_y)

z = y**3 - 2
dz_dy = diff(z)

dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

print(dz_dx_chain)
print(dz_dx_no_chain)
```

```python
# a = min of the x range
# b = max of the x range
# n = number of rectangles
# f = function we are integrating

def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
        
    return total_sum * delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)

print(area)
```

```python
from sympy import *

x, i, n = symbols('x i n')

f = x**2 + 1
lower, upper = 0, 1

delta_x = ((upper - lower) / n) # determine the length
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

area = limit(n_rectangles, n, oo)

print(float(area))
```

```python
from sympy import *

# Declare variables to SymPy
x, i, n = symbols('x i n')

# Declare function and range
f = x**2 + 1
lower, upper = 0, 1

# Calculate width and each rectangle height at index 'i'
delta_x = ((upper - lower) / n) # length of each rectangle
x_i = (lower + delta_x * i)     # Start of each rectangle, i is the index of each rectangle
fx_i = f.subs(x, x_i)           # Height of each rectangle at index i

# Iterate all 'n' rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# Calculate the area by approaching the number of rectangles 'n' to infinity
area = limit(n_rectangles, n, oo)

print(float(area))
```
